```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import numpy as np
import seaborn as sns
from collections import namedtuple
from scipy.optimize import curve_fit
```

```python
def plot_active_fraction(active_data, error_data, conditions, groups,):
    ax = sns.barplot(x=groups, y=active_data, hue=conditions, palette=sns.color_palette("colorblind"))
    x_coords = [p.get_x() + 0.5*p.get_width() for p in ax.patches]
    y_coords = [p.get_height() for p in ax.patches]
    ax.errorbar(x=x_coords, y=y_coords, yerr=error_data, fmt="none", c= "k")
    ax.set_ylabel("Inactive Fraction")
    ax.set_title("Inactive alleles over 16h")
    plt.show()
```
